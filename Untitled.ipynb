```python
import geopandas as gpd 
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt  
import os
import folium
```

```python
gdf = gpd.read_file("./Daten/geodata/canton_map.gdb")
tree = pd.read_csv("./Daten/TREE2_Data_IWI_Open_Data_Vorlesung_2023_label.csv")
tree = tree.dropna(how = 'any', axis = 0)
cantons = pd.read_csv("./Daten/cantons_codes.csv")

tree_canton = tree.merge(cantons,
                         how="left",
                         on='aes_canton')

data = gdf.merge(tree_canton,
                        how="left", 
                        on='NAME',
                        indicator=True)

geo_data = gpd.GeoDataFrame(data)
```

```python
def map_data(data, title):
    fig, ax = plt.subplots(figsize=(10, 10))

    # Set the t0fmedu_comp minimum and maximum, so the coloring of the data makes more sense.
    #vmin = data['t0fmedu_comp'].min()
    #vmax = data['t0fmedu_comp'].quantile(0.95)

    # Create the Geo Plot on the column.
    data.plot(column='t0fmedu_comp',
              cmap='Spectral',
              ax=ax,
              legend=True,
              missing_kwds={'color': 'lightgrey', 'edgecolor': 'red', 'hatch': '///'})

    ax.set_title(title)
    

    plt.show()

map_data(geo_data, 'Parents\' Highest Educational Attainment')
```
