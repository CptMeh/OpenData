```python
import geopandas as gpd 
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt  
import os
```

```python
tree = pd.read_csv(".\Daten\TREE2_Data_IWI_Open_Data_Vorlesung_2023_label.csv")
tree = tree.dropna(how = 'any', axis = 0)
#tree
```

```python
cantons = pd.read_csv(".\Daten\cantons_codes.csv")
#cantons
```

```python
tree_canton = tree.merge(cantons,
                         how="left",
                         on='aes_canton')
tree_canton
```

```python
data_map = gdf.merge(tree_canton,
                        how="left", 
                        on='NAME',
                        indicator=True)
#data_map
```

```python
fig, ax = plt.subplots(figsize=(10,10))

# Set the variable minimum and maximum, so the coloring of the data makes more sence.
vmin = data_map['t1wt'].min()
vmax = data_map['t1wt'].quantile(0.95)

# Create the Geo Plot on the column.
data_map.plot(column='t1wt', 
                cmap='RdYlGn', 
                ax=ax, 
                legend=True, 
                missing_kwds={"color": "lightgrey", "edgecolor": "red", "hatch": "///"},
                scheme="quantiles",
                vmin=vmin, 
                vmax=vmax)    

ax.set_title('Title')
```
