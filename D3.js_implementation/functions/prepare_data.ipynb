```python
import pandas as pd
import numpy as np
import os
```

```python
tree = pd.read_csv("../Daten/TREE2_Data_IWI_Open_Data_Vorlesung_2023_label.csv")
indeces = pd.read_csv("../Daten/cantons_indeces.csv")
indeces.set_index("aes_canton", inplace=True)
indeces = indeces.to_dict()["index"]

tree['aes_canton'] = tree['aes_canton'].map(indeces)
tree = tree.replace("NaN", np.NaN)
tree = tree.dropna(how="any", axis=0)
tree
```

```python
#write to currated_data.csv in Daten
os.chdir("../Daten")
tree.to_csv("currated_data.csv", encoding='utf-8', index=False)
```
