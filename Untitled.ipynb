```python
import geopandas as gpd 
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt  
import os
```

```python
tree = pd.read_csv(".\Daten\TREE2_Data_IWI_Open_Data_Vorlesung_2023_label.csv")
tree = tree.dropna(how = 'any', axis = 0)
tree
```

```python
cantons = pd.read_csv(".\Daten\cantons_codes.csv")
cantons
```

```python
tree_canton = tree.merge(cantons,
                         how="left",
                         on='aes_canton')
tree_canton
```

```python
data_map = gdf.merge(tree_canton,
                        how="left", 
                        on='NAME',
                        indicator=True)
data_map
```

```python
fig, ax = plt.subplots(figsize=(10,10))

grouped_df = data_map.groupby('NAME')['t0hisei08_3q'].value_counts(normalize=True).mul(100).rename('percentage_t0hisei08_3q_high').reset_index()

# Specify the specific answer you are interested in
specific_answer = 'High'

# Filter the grouped DataFrame for the specific answer
filtered_df = grouped_df[grouped_df['t0hisei08_3q'] == specific_answer]

# Print the result
print(filtered_df)

df = data_map.merge(filtered_df,
                        how="left",
                        on='NAME')

# Set the variable minimum and maximum, so the coloring of the data makes more sence.
#vmin = data_map['t0hisei08_3q'].min()
#vmax = data_map['t0hisei08_3q '].quantile(0.95)

# Create the Geo Plot on the column.
df.plot(column='percentage_t0hisei08_3q_high',
                cmap='RdYlGn', 
                ax=ax, 
                legend=True, 
                missing_kwds={"color": "lightgrey", "edgecolor": "red", "hatch": "///"}
               )

ax.set_title('Title')
```
